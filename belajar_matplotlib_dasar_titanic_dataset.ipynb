- Second Class : 12

## Mengenal Log Scale

```python
fig, ax = plt.subplots()
_ = ax.set_title('Titanic Data Set')
_ = ax.plot(np.sort(df['fare']), marker='o', markersize=1, linestyle='')
_ = ax.set_ylabel('Fares')
_ = ax.set_xlabel('Passengers')

_ = ax.set_yscale('log')
_ = ax.set_yticks([3, 12, 30, 870])
_ = ax.set_yticklabels(['3rd', '2rd', '1stb', '1sts'])
```

## Mengenal Mutiple Axes [1/3]

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
_ = fig.suptitle('Super Title')
_ = ax1.set_title('Judul Axes 1')
_ = ax2.set_title('Judul Axes 2')
```

- Sebuah `Figure` dapat memiliki lebih dari satu `Axes`
- `ncols` digunakan untuk mengatur jumlah kolom
- `(ax1, ax2)` digunakan untuk menampung kedua objects `Axes`
- `fig.suptitile()` digunakan untuk mengatur judul `Figure`
- `ax1.set_title()` digunakan untuk mengatur judul `Axes`

## Mengenal Multiple Axes [2/3]

```python
fig, (ax1, ax2) = plt.subplots(nrows=2)
_ = fig.suptitle('Super Title')
```

## Plotting pada Multiple Axes

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
_ = fig.suptitle('Titanic Data Set')

_ = ax1.plot(np.sort(df['age']), marker='o', markersize=1, linestyle='')
_ = ax1.set_title('Ages')
_ = ax1.set_xlabel('Passengers')
# _ = ax1.set_ylabel('Ages')
_ = ax1.set_yticks([15, 25, 55, 64])
_ = ax1.set_yticklabels(['children', 'youth', 'adults', 'seniors'])

_ = ax2.plot(np.sort(df['fare']), marker='o', markersize=1,  linestyle='')
_ = ax2.set_title('Fares')
_ = ax2.set_xlabel('Passengers')
# _ = ax2.set_ylabel('Fares')
_ = ax2.set_yscale('log')
_ = ax2.set_yticks([3, 12, 30,  870])
_ = ax2.set_yticklabels(['3rd', '2rd', '1stb', '1sts'])
```

## Mengenal Constained Layout

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, constrained_layout=True)
_ = fig.suptitle("Titanic Data Set")

_ = ax1.plot(np.sort(df['age']), marker='o', markersize=1, linestyle='')
_ = ax1.set_title('Ages')
_ = ax1.set_xlabel('Passengers')
_ = ax1.set_ylabel('Ages')
_ = ax1.set_yticks([15, 25, 55, 64])
_ = ax1.set_yticklabels(['children', 'youth', 'adults','seniors'])

_ = ax2.plot(np.sort(df['fare']), marker='o', markersize=1, linestyle='')
_ = ax2.set_title('Fares')
_ = ax2.set_xlabel('Passengers')
_ = ax2.set_ylabel('Fares')
_ = ax2.set_yscale('log')
_ = ax2.set_yticks([3, 12, 30, 870])
_ = ax2.set_yticklabels(['3rd', '2rd', '1stb', '1sts'])
```

- `constrained_layout` digunakan untuk memastikan tidak ada element dalam plotting yang tumpang tindih (overlap)
- [Referensi `Constrained Layout`](https://matplotlib.org/3.1.1/tutorials/intermediate/constrainedlayout_guide.html)

## Mengenal GridSpec

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, constrained_layout=True, gridspec_kw={'width_ratios':[1,3]})
_ = fig.suptitle('Titanic Data Set')

_ = ax1.plot(np.sort(df['age']), marker='o', markersize=1, linestyle='')
_ = ax1.set_title('Ages')
_ = ax1.set_xlabel('Passengers')
_ = ax1.set_ylabel('Ages')
_ = ax1.set_yticks([15, 25, 55, 64])
_ = ax1.set_yticklabels(['children', 'youth', 'adults', 'seniors'])

_ = ax2.plot(np.sort(df['fare']), marker='o', markersize=1, linestyle='')
_ = ax2.set_title('Fares')
_ = ax2.set_xlabel('Passengers')
_ = ax2.set_ylabel('Fares')
_ = ax2.set_yscale('log')
_ = ax2.set_yticks([3, 12, 30, 870])
_ = ax2.set_yticklabels(['3rd', '2rd', '1stb', '1sts'])
```

## Mengenal Line Plot

```python
x = np.linspace(0, 2*np.pi, 100)
sin_x = np.sin(x)

fig, ax= plt.subplots()
_ = ax.plot(x, sin_x)
```

## Mengenal Aspect Ratio

```python
fig, ax = plt.subplots()
_ = ax.plot(x, sin_x)
_ = ax.set_aspect('equal')
```

## Mengenal Marker

```python
fig, ax = plt.subplots(figsize=(10,5))
_ = ax.plot(x, sin_x, marker='*')
_ = ax.set_aspect('equal')
```

[Referensi `Marker`](https://matplotlib.org/examples/lines_bars_and_markers/marker_reference.html)

## Pengaturan Warna

```python
fig, ax = plt.subplots(figsize=(10,5))
_ = ax.plot(x, sin_x, marker='*', markerfacecolor='white', markeredgecolor='blue', color='red')
_ = ax.set_aspect('equal')
```

- `markerfacecolor` digunakan untuk mengatur warna marker
- `markeredgecolor` digunakan untuk mengatur warna garis tepi marker
- `color` digunakan untuk mengatur warna plot utama, dalam kasus ini adalah garis (line)

## Pengaturan Ukuran

```python
fig, ax = plt.subplots(figsize=(10,5))
_ = ax.plot(x, sin_x, marker='*', markersize=20, linewidth=20)
_ = ax.set_aspect('equal')
```

- `markersize` digunakan untuk mengatur ukuran marker
- `linewidth` digunakan untuk mengatur ketebalan garis (line)

## Mengenal Line Style

```python
fig, ax = plt.subplots(figsize=(10,5))
_ = ax.plot(x, sin_x, linewidth=5, linestyle='-.')
_ = ax.set_aspect('equal')
```

[Referensi `Line Style`](https://matplotlib.org/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py)

## Mengenal Legend

```python
fig, ax = plt.subplots(figsize=(10,5))
_ = ax.plot(x, sin_x, markersize=20, linestyle='-.', color='red', label='sin')
_ = ax.plot(x, np.cos(x), markersize=20, linestyle='-', color='blue', label='cos')
_ = ax.set_aspect('equal')
_ = ax.legend()
```

[Referensi `Legend`](https://matplotlib.org/3.1.0/tutorials/intermediate/legend_guide.html)

## Mengenal `ListedColormap`

```python
import matplotlib.colors as mcolors

cmap = mcolors.ListedColormap(['red', 'green'])

fig, ax = plt.subplots()

_ = ax.scatter('age', 'fare', c='survived', cmap=cmap, data=df)

_ = ax.set(xlabel='age', ylabel='fare', title='Titanic Dataset')
```

## Menyertakan Legend

```python
fig, ax = plt.subplots()
sc = ax.scatter('age', 'fare', c='survived', cmap=cmap, data=df)
_ = ax.set(xlabel='Age', ylabel='Fare')

_ = ax.legend(*sc.legend_elements())
```

## Mengenal Custom Legend

```python
fig, ax = plt.subplots()
sc = ax.scatter('age', 'fare', c='survived', cmap=cmap, data=df)
_ = ax.set(xlabel='Age', ylabel='Fare')

_ = ax.legend(handles=sc.legend_elements()[0], labels=['Died', 'Survived'])
```

## Pengaturan `facecolor` dan `edgecolor`

```python
fig, ax = plt.subplots()

custom_bins = [0, 15, 25, 55, 64]
(counts, egdes, _) = ax.hist('age', bins=custom_bins, data=df, facecolor='white', edgecolor='black')

_ = ax.set_xticks(custom_bins)
```

## Mengenal `groupby`

```python
fig, ax = plt.subplots()
custom_bins = [0, 15, 25, 55, 64]

for label, gdf in df.groupby('sex'):
    ax.hist('age', bins=custom_bins, data=gdf, label=label, alpha=.5)
    
_ = ax.set_xticks(custom_bins)
_ = ax.legend()
```

## Mengenal `histtype`

```python
fig, ax = plt.subplots()
custom_bins = [0, 15, 25, 55, 64]

for label, gdf in df.groupby('sex'):
    ax.hist('age', bins=custom_bins, data=gdf, label=label, alpha=.5, histtype='step')
    
_ = ax.set_xticks(custom_bins)
_ = ax.legend()
```
